# Mapa de calor correlacion entre distintas variables (opcion1)


```python
fifa20_2 = fifa20[["age", "height_cm", "weight_kg", "overall", "potential", "value_eur", "wage_eur", "power_jumping", "power_long_shots", "skill_moves", "shooting", "passing", "dribbling", "defending", "physic", "movement_acceleration", "movement_agility", "mentality_vision"]]
```

```python
import seaborn as sns

f, ax = plt.subplots(figsize=(10, 8))
corr = fifa20_2.corr()
sns.heatmap(corr)
```

# Mapa de calor correlacion entre distintas variables (opcion2)


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = fifa20_2.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=.5)
plt.title("Mapa de Calor de Correlación entre Variables")
plt.show()
```
